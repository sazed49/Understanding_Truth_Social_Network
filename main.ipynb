Loading "external_urls" table

```python
import pandas as pd

external_urls = pd.read_csv('S:/Fall 24/SNA/Project/truth_social/Dataset/external_urls.tsv', sep='\t')

external_urls
```

Loading "truth_external_url_egdes" table

```python
truth_external_url_edges = pd.read_csv('S:/Fall 24/SNA/Project/truth_social/Dataset/truth_external_url_edges.tsv', sep='\t')

truth_external_url_edges
```

```python
print(truth_external_url_edges.columns)      # Check columns in truth-external_url_edges
print(external_urls.columns)
```

making a table named " truth_and_url" joining "truth_external_url_edges" and "external_urls" with following columns:
truth_id , url_id, url

```python
truth_and_url = truth_external_url_edges.merge(external_urls, left_on="url_id", right_on="id", suffixes=('', '_y'))

# Select the desired columns
truth_and_url = truth_and_url[['truth_id', 'url_id', 'url']]

print(truth_and_url)
```

Finding author id joining previous table and "truths" table   to form a new table "truthID_authorId_url" with following columns:
truth_id, author_id, url,

```python
# Merge last table with `truth` to add the author_id
truthID_authorId_url = truth_and_url.merge(truths[['id', 'author']], left_on="truth_id", right_on="id", suffixes=('', '_y'))

# Select only the desired columns
truthID_authorId_url = truthID_authorId_url[['truth_id', 'author',  'url']]

# Rename columns for clarity
truthID_authorId_url = truthID_authorId_url.rename(columns={'author': 'author_id', 'url': 'url'})

print(truthID_authorId_url)
```

Join previous table with "users" table to get the username for corresponding user_id.
new table - "truth_sername_url"
columns -truth_id,user_id,username,url

```python
# Merge last table with `authors` to add the username
truth_username_url = truthID_authorId_url.merge(users[['id', 'username']], left_on="author_id", right_on="id", suffixes=('', '_y'))

# Select only the desired columns
truth_username_url = truth_username_url[['truth_id', 'author_id', 'username', 'url']]

# Rename 'author_id' to 'user_id' for clarity
truth_username_url = truth_username_url.rename(columns={'author_id': 'user_id'})

print(truth_username_url)
```

Checking if any NaN values. But there is no. Which means joining was perfect!

```python
# Count NaN values in each column of merged_result
nan_counts = truth_username_url.isna().sum()

print(nan_counts)
```

Getting the domain names from the last table

```python
# extract the domain from the url
from urllib.parse import urlparse
truth_username_url['domain'] = truth_username_url['url'].apply(lambda x: urlparse(x).netloc)
truth_username_url
```

```python
# find the number of unique domains
truth_username_url['domain'].nunique()
```
